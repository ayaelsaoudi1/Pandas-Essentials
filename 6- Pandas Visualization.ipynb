# Pandas Visualization

This notebook covers various visualization techniques using Pandas, from line and bar plots to scatter, histogram, and pie charts. These visualizations help in understanding data trends, distributions, and relationships.

---

## Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

We import:
- `pandas` for handling data,
- `numpy` for numerical operations,
- `matplotlib.pyplot` for plotting customization.

---



```python
df = pd.read_csv(r"dummy data\Ice Cream Ratings.csv")
df = df.set_index('Date')  # Setting the Date column as the index
df
```

## Loading the Dataset

- The dataset is loaded from a CSV file.
- The `Date` column is set as the index for time-series visualization.

---

## Default Line Plot


```python
df.plot()  # Default is a line plot
```

- This generates a simple line plot for all numeric columns.


```python
df.plot(kind='line')
```

- This explicitly specifies a line plot.

---

## Available Matplotlib Styles


```python
print(plt.style.available)
```

- This prints all available Matplotlib styles for customization.



```python
plt.style.use('classic')
```

- This sets the plot style to `classic` for better aesthetics.

---

## Line Plots with Customizations


```python
df.plot(kind='line', subplots=True)
```

- This creates multiple line plots, one for each column, in separate subplots.


```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

- This adds a title, x-label, and y-label to the line plot.

---

## Bar Charts

```python
df.plot(kind='bar')
```

- This generates a vertical bar chart.

```python
df.plot(kind='bar', stacked=True)
```

- This creates a stacked bar chart.

```python
df['Flavor Rating'].plot(kind='bar')
```

- This plots a bar chart for a single column (`Flavor Rating`).

```python
df.plot.barh(stacked=True)
```

- This creates a horizontal stacked bar chart.

---

## Scatter Plots

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating')
```

- This generates a scatter plot of `Texture Rating` vs. `Overall Rating`.

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='Red')
```

- This adjusts the scatter plot:
  - `s=500`: Increases point size.
  - `c='Red'`: Colors points red.

---

## Histograms

```python
df.plot.hist()
```

- This creates a histogram with default bins.


```python
df.plot.hist(bins=20)
```

- This increases the number of bins to 20 for better granularity.

---

## Boxplots

```python
df.boxplot()
```

- This generates a boxplot to visualize data distribution and outliers.

---

## Area Plot

```python
df.plot.area(figsize=(10,5))
```

- This creates an area plot with a specified figure size.

---

## Pie Chart



```python
df.plot.pie(y='Flavor Rating')
```

- This generates a pie chart for `Flavor Rating`.
